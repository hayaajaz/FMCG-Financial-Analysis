```python
import pandas as pd


dataset = pd.read_excel(r"C:\Users\hoora\OneDrive\Desktop\Project 2\FMCG_Financial_Performance.xlsx")
all_sheets = pd.read_excel(
    r"C:\Users\hoora\OneDrive\Desktop\Project 2\FMCG_Financial_Performance.xlsx",
    sheet_name=None
)

sales = all_sheets["Actual_Sales_Transactions"]
cost = all_sheets["Standard_Cost_Master"]
promo = all_sheets["Promotion_Budget"]
salary = all_sheets["Employee_Salary"]
```

```python
import pandas as pd
import re

# 1. Define the Cleaning Function
# This uses 'Regex' to strip all non-numeric characters (PKR, Rs, /-) in one go
def fix_money(value):
    if pd.isna(value):
        return 0.0
    # Keep only numbers, dots, and minus signs
    clean_string = re.sub(r'[^\d.-]', '', str(value))
    try:
        return float(clean_string)
    except:
        return 0.0

# 2. Fix the 'Sales' sheet (The messy one)
# Strip currency from both money columns
sales['Gross_Sales_Value'] = sales['Gross_Sales_Value'].apply(fix_money)
sales['Trade_Discount_Given'] = sales['Trade_Discount_Given'].apply(fix_money)

# Standardize Dates
sales['Date'] = pd.to_datetime(sales['Date'])

# Clean SKU codes (remove invisible spaces)
sales['Global_SKU_Code'] = sales['Global_SKU_Code'].str.strip().str.upper()

# 3. Fix the 'Cost' sheet
cost['Global_SKU_Code'] = cost['Global_SKU_Code'].str.strip().str.upper()

# 4. Save everything back to ONE Excel file with 4 Tabs
# This creates a file exactly like your raw data but with 'Numbers' instead of 'Text'
with pd.ExcelWriter('Cleaned_Master_FMCG.xlsx', engine='xlsxwriter') as writer:
    sales.to_excel(writer, sheet_name='Actual_Sales_Transactions', index=False)
    cost.to_excel(writer, sheet_name='Standard_Cost_Master', index=False)
    promo.to_excel(writer, sheet_name='Promotion_Budget', index=False)
    salary.to_excel(writer, sheet_name='Employee_Salary', index=False)

print("Process Complete! Your clean file is saved as 'Cleaned_Master_FMCG.xlsx'")
```
